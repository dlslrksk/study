```python
# 1. 그룹화 국가

select_sql_3 = """
                SELECT * FROM sales 
                GROUP BY Country
            """

result_3 = _db.executeAll(select_sql_3)
dt_3 = pd.DataFrame(result_3)
dt_3
```

```python
# 2. 출력 >> Country, Countrty 갯수를 구하는 dataframe 만들기

select_sql_3 = """
                SELECT Country, COUNT(`Country`) AS cnt
                FROM sales
                GROUP BY Country
            """

result_3 = _db.executeAll(select_sql_3)
dt_3 = pd.DataFrame(result_3)
dt_3
```

```python
# 3. Country 개수로 내림차순 정렬

select_sql_3 = """
                SELECT Country, COUNT(`Country`) AS cnt
                FROM sales
                GROUP BY Country
                ORDER BY cnt ASC
            """

result_3 = _db.executeAll(select_sql_3)
dt_3 = pd.DataFrame(result_3)
dt_3
```
