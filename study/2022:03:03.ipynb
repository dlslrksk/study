```python
corona_df = pd.read_csv('./pandas/corona.csv')
corona_df.columns = ['인덱스', '등록일시', '사망자', '확진자', '게시글번호', '기준일', '기준시간', '수정일시', '누적_의심자', '누적_확진율']
corona_df_2 = corona_df.sort_values("등록일시")
corona_df_2["일일_확진자"] = (corona_df_2["확진자"] - corona_df_2["확진자"].shift()).fillna(0)
corona_df_2["일일_사망자"] = (corona_df_2["사망자"] - corona_df_2["사망자"].shift()).fillna(0)
corona_df_3 = corona_df_2.drop(['인덱스','기준일','기준시간','수정일시','게시글번호'], axis=1)
corona_df_3.reset_index(drop=True, inplace=True)
corona_df_3.fillna(0)
```

```python
# 코로나 누적_확진율이 3.5이상이면 'High', 3.5미만이면 'Low'

corona_df_5 = corona_df_3.copy()
corona_df_5['H/L'] = corona_df_5['누적_확진율'].apply(lambda x : 'Low' if (x < 3.5) else 'High')
corona_df_5.fillna(0)
```

```python
# pd.to_datetime >> 날짜형으로 변환
# .dt.weekday >> 요일로 변환

corona_df_5['등록일시'] = pd.to_datetime(corona_df_5['등록일시'], format = "%Y-%m-%d")

corona_df_5['요일'] = corona_df_5['등록일시'].dt.weekday
corona_df_5.head()
```
