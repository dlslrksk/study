```python
import scipy.stats as stats
```

```python
# stats.norm(평균, 표준편차) >> 가우시안 정규분포 객체를 생성하는 메서드
# https://namyoungkim.github.io/scipy/probability/2017/09/04/scipy/

y_ = stats.norm(0,1).pdf(x)
y_
```

```python
plt.plot(x,y_)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def norm(_a,_b,_x):
    return stats.norm(_a,_b).pdf(_x)

x = np.linspace(-20,20,101)
x
```

```python
for i in range(1,11,1):
    _y = norm(0,i,x)
    plt.plot(x,_y)
plt.show()
```

```python
# 파이차트

plt.style.use('default') # 스타일지정 // https://hong-yp-ml-records.tistory.com/88

x = np.random.randint(10,50,5)
labels = ['A','B','C','D','E'] # 이름표 붙이기
explodes = [0.1,0,0,0,0]
plt.pie(x, autopct='%.1f%%', labels=labels, counterclock=False, startangle=180, explode=explodes, textprops = dict(color="BLACK")) 
# counterclock >> 시계방향변경
# startangle >> 시작점
# explode >> 거리 띄우기
# textprops = dict(color="") >> 글자색 바꾸기
plt.show()
```

```python
# 선형그래프
# x, y 개수가 동일해야 함

x = np.linspace(1,10,10)
y = np.random.rand(10)

plt.plot(x,y)
plt.xticks(np.arange(min(x),max(x)+1,1.0))
plt.yticks(np.arange(min(y),max(y)+0.1,0.1).round(2))
plt.xlabel('X')
plt.ylabel('Y')
plt.legend('X', loc = 0) # 범례, loc = 숫자 >> 범례위치
plt.grid() # 격자
plt.show()
```
