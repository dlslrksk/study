```python
# emp collection
# key 값이 EMPNO, ENAME, DEPTNO만 출력이 되도록 코드를 작성

# SELECT EMPNO, ENAME, DEPTNO FROM emp

# list(db.emp.find(
#     {},           ## WHERE절 : 조건을 쓸 때 사용하는 공간
#     {}            ## COLUMN절 : key 값을 선택할 수 있는 공간
#     ))

result = db.emp.find(
        {}, 
        {'EMPNO':1, 'ENAME':1, 'DEPTNO':1}
        )

# for list in result:
#     print(list)

pd.DataFrame(result)
```

```python
# SELECT * FROM emp ORDER BY ENAME
# sort("ENAME", 1) >> 오름차순
# sort("ENAME", -1) >> 내림차순

result = db.emp.find().sort("ENAME")

# for list in result:
#     print(list)

pd.DataFrame(result)
```

```python
# SELECT * FROM emp WHERE SAL > 2450
# 조건에 해당하는 값이 숫자인지 문자인지 확인
# {'$gt':2450}, >2450
# {'$gte':2450}, >=2450
# {'$lt':2450}, <2450
# {'$lte':2450}, <=2450

result = db.emp.find(
    {
        'SAL' : {'$gt': 2450}
    }
    )

# for list in result:
#     print(list)

pd.DataFrame(result)
```

```python
# SELECT * FROM emp WHERE ENAME = SMTIH OR ENAME = ALLEN

result = db.emp.find(
    {
        '$or' : 
        [{"ENAME" : "SMITH"},
        {"ENAME" : "ALLEN"}]
    }
    )

# for list in result:
#     print(list)

pd.DataFrame(result)
```

```python
# SELECT * FROM emp WHERE ENAME = SMTIH AND JOB = CLERK

result = db.emp.find(
    {
        "ENAME" : "SMITH",
        "JOB" : "CLERK"
    }
    )

# for list in result:
#     print(list)

pd.DataFrame(result)
```

```python
# SELECT * FROM emp WHERE SAL >= 2000 AND SAL < 4000

result = db.emp.find(
    {
        '$and' : 
        [{'SAL' : {'$gte': 2000}},
        {'SAL' : {'$lt': 4000}}]
    }
    )

# for list in result:
#     print(list)

pd.DataFrame(result)
```

```python
# SELECT * FROM emp WHERE SAL >= 2000 AND SAL < 4000
# ',' 좀 잘 적어라!!!

result = db.emp.find(
    {
        'SAL' : {'$gte': 2000, '$lt': 4000}
    },
    {
        '_id' : 0
    }
    )

# for list in result:
#     print(list)

pd.DataFrame(result)
```

```python
# SELECT * FROM emp WHERE ENAME LIKE %S%
# '$regex' : 'S' >> 포함
# '$regex' : '^S' >> 시작
# '$regex' : 'S$' >> 끝

result = db.emp.find(
    {
        'ENAME' : {
            '$regex' : 'S'
            }
    },
    {
        '_id' : 0
    }
    )

# for list in result:
#     print(list)

pd.DataFrame(result)
```

```python
# SELECT ENAME, JOB, SAL, DEPTNO FROM emp 
# WHERE SAL > 1250 AND SAL <= 3000 AND JOB = SALESMAN
# ORDER BY ENAME

result = db.emp.find(
    {
        'SAL' : {'$gt': 1250, '$lte': 3000},
        'JOB' : 'SALESMAN'
    },
    {
        'ENAME':1, 'JOB':1, 'SAL':1, 'DEPTNO':1, '_id':0
    }
    ).sort("ENAME", 1)

# for list in result:
#     print(list)

pd.DataFrame(result)
```
