```python
import FinanceDataReader as fdr
#KOSPI(KS11)
df = fdr.DataReader('KS11', "2022")
df['Close'].plot()
```

```python
import ipywidgets as widgets 

from ipywidgets import interact, interact_manual 

import pandas as pd

import numpy as np 

import matplotlib.pyplot as plt 

%matplotlib inline 
```

```python

close_df = pd.DataFrame({'kospi_close':df['Close']}) 

close_df.tail(50)
```

```python
logret_df = np.log(close_df / close_df.shift(1)) 

logret_df.tail(50)
```

```python

logret_df = logret_df.dropna() 

logret_df.columns = ['Close']

logret_df.tail(50)
```

```python
# import FinanceDataReader as fdr

# df = fdr.DataReader('USD/KRW', '2010')
# df.tail(50)
```

```python
mu = logret_df.mean() 

mu
```

```python
# 만약 년간 기대수익률을 추정하고 싶다면 위에 계산된 일별 로그수익률 mu에 단순히 1년간 거래일(여기서는 252일을 사용합니다)을 곱해주면 됩니다.
# 이는 기대수익률 추정에 로그수익률을 사용하였기 때문입니다.

mu = logret_df.mean() * 252

mu
```

```python
logret_df.std()

# 연 환산

logret_df.std() * np.sqrt(252)
```

```python
mu = mu.values 

mu
```

```python
# 공분산

logret_df.cov()
```

```python
# cov_mat : 공분산 행렬(Covariance Matrix)

cov_mat = logret_df.cov() * 252 

cov_mat = cov_mat.values # 행렬구조로 저장합니다 

cov_mat
```

```python
# 주식 100%

w = np.array([1]) 

# port_return: 포트폴리오의 기대수익률 

port_return = w.dot(mu)

port_return
```

```python
# 포트폴리오의 Risk(=표준편차)를 계산합니다. 

port_var = np.dot(np.dot(w, cov_mat), w.T) 

port_std = np.sqrt(port_var) 

print("Portfolio Risk: ", port_std)
```
