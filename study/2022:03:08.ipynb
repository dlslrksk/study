```python
import pandas as pd
```

```python
class Class_1():

    def __init__(self, input_url):
        self.sales_df = pd.read_csv('./csv/'+input_url)
        self.drop_list = []


    def _sort(self, input_sort = 'Country'):
        self.sales_df.sort_values(input_sort, inplace=True)

        self.sales_df.reset_index(drop=True, inplace=True)

        return self.sales_df
    
    def _drop(self, *drop_columns): # *변수 : 끝을 정확히 모르는 변수 전체
        for list in drop_columns: # drop_columns에 입력된 인자값을 list로 추출
            self.drop_list.append(list) # 추출된 list를 self.drop_list에 append
        self.sales_df = self.sales_df.drop(self.drop_list, axis=1) # self.drop_list의 내용을 drop

        return self.sales_df

# for list in list_array
# 처음 변수를 지정할 때 나비효과를 생각하자

dt = Class_1('Sales Records.csv')

dt.sales_df

dt._sort()

# dt._drop('Order ID','Ship Date') # 두 개가 안되는 이유 물어보기
```
