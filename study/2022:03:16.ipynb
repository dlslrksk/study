```python
# CASE

select_sql_16 = """
                SELECT ENAME,
                CASE
                    WHEN JOB = 'ANALYST' THEN '분석가'
                    WHEN JOB = 'CLERK' THEN '점원'
                    WHEN JOB = 'MANAGER' THEN '관리자'
                    WHEN JOB = 'PRESIDENT' THEN '대통령'
                    WHEN JOB = 'SALESMAN' THEN '판매사원'
                END as JOB
                FROM emp   
            """

result_16 = _db.executeAll(select_sql_16)
dt_16 = pd.DataFrame(result_16)
dt_16
```

```python
# 1. emp table과 dept table를 DEPTNO을 기준으로 left join
# 2. ENAME, JOB, SAL, DEPTNO, LOC 출력
# 3. ENAME, LOC는 그래도 출력
# 4. JOB은 각각 한글명으로 변경
# 5. SAL은 2450을 기준으로 2450이상이면 HIGH, 2450미만이면 LOW
# 6. DEPTNO은 DEPTNO의 값과 dept table의 DNAME 값을 더해서 표시(DEPTNO-DNAME 형식)
# 7. 모든 컬럼의 이름은 한글로 변경
```

```python
select_sql_17 = """
                SELECT
                emp.ENAME as '사원명',
                CASE
                    WHEN emp.JOB = 'ANALYST' THEN '분석가'
                    WHEN emp.JOB = 'CLERK' THEN '점원'
                    WHEN emp.JOB = 'MANAGER' THEN '관리자'
                    WHEN emp.JOB = 'PRESIDENT' THEN '회장'
                    WHEN emp.JOB = 'SALESMAN' THEN '판매사원'
                END as '직급',
                IF(emp.SAL >= 2450, 'HIGH', 'LOW') as '월급',
                CONCAT(emp.DEPTNO, '-', dept.DNAME) as '부서번호-부서명',
                dept.LOC as '위치'
                FROM emp left join dept
                ON emp.DEPTNO = dept.DEPTNO
                ORDER BY `사원명` ASC
                """

result_17 = _db.executeAll(select_sql_17)
dt_17 = pd.DataFrame(result_17)
dt_17
```

```python
# 부서의 이름이 'sales'인 소속 사원들의 리스트를 출력

select_sql_20 = """
                SELECT 
                emp.ENAME, 
                dept.DNAME
                FROM emp left join dept
                ON emp.DEPTNO = dept.DEPTNO
                WHERE dept.DNAME = 'sales'
            """

result_20 = _db.executeAll(select_sql_20)
dt_20 = pd.DataFrame(result_20)
dt_20
```

```python
# 부서의 이름이 'sales'인 소속 사원들의 리스트를 출력
# 이중쿼리문

select_sql_21 = """
                SELECT 
                emp.ENAME,
                dept.DNAME        
                FROM emp, dept
                WHERE emp.DEPTNO = dept.DEPTNO
                AND dept.DEPTNO =
                (
                    SELECT dept.DEPTNO FROM dept
                    WHERE DNAME = "sales"
                )

            """

result_21 = _db.executeAll(select_sql_21)
dt_21 = pd.DataFrame(result_21)
dt_21
```

```python
# 부서의 위치가 'NEW YORK', 'DALLAS'가 아닌 사원들의 리스트를 출력
# 이중쿼리문

select_sql_22 = """
                SELECT 
                emp.ENAME,
                dept.LOC
                FROM emp left join dept
                ON emp.DEPTNO = dept.DEPTNO
                WHERE dept.DEPTNO IN
                (
                    SELECT dept.DEPTNO 
                    FROM dept
                    WHERE LOC 
                    NOT IN ('NEW YORK', 'DALLAS')
                )

            """

result_22 = _db.executeAll(select_sql_22)
dt_22 = pd.DataFrame(result_22)
dt_22
```
