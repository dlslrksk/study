```python
def func_1(a,b):
    c = a + b
    return c

d = func_1(3,6)

print(d)
```

```python
def func_2(s_num, l_num, a_num):
    sum = 0
    for i in range(s_num, l_num, a_num):
        sum += i
    return sum

sum = func_2(1,11,1)
print(sum)
```

```python
def func_4(*args):
    result = 0 
    for i in args:
        result += i
    return result

result = func_4(1,2,3,4,5)
print(result)

result2 = func_4(1,2,3,4,5,6,7,8,8,9,10)
print(result2)

result3 = func_4(2,4,6,8,10)
print(result3)
```
