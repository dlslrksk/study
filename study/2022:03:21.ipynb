```python
# 4. mysql query 데이터 받아오기(Module 생성, Class로 구현)

import pymysql
from modules import mod_sql

_db = mod_sql.Database()

select_sql = '''
                SELECT *
                FROM apple
            '''

result = _db.executeAll(select_sql)
dt = pd.DataFrame(result)
dt

```

```python
# 5. 주식의 기준 >> 시가, 종가, 고가, 저가의 값을 반올림하여 소수점 2번째 자리까지 나타내기

round_sql = '''
                SELECT open, close, high, low
                FROM apple
            '''

result = _db.executeAll(round_sql)
dt = pd.DataFrame(result)
dt.round(2)

# round_sql = '''
#                 SELECT open, close, high, low
#                 FROM apple LIMIT 1000
#             '''

# result = _db.executeAll(round_sql)
# dt = pd.DataFrame(result).round(2)
```

```python
# 6. 주식 기준 >> NAN값이 존재하는 경우 최초의 값으로 변환

apple_df = apple_df.fillna(method='bfill').round(2)

apple_df

# NAN값 확인 dt.isna().sum()
```

```python
# 7. 컬럼 추가 >> 등락폭(종가-시가) 값을 삽입

apple_df["등락폭(종가-시가)"] = (apple_df["Close"] - apple_df["Open"]).fillna(0)

apple_df
```

```python
# 8. 컬럼 추가 >> 등락폭/시가 * 100% 삽입

apple_df["등락률"] = apple_df["등락폭(종가-시가)"]/apple_df["Open"]*100

apple_df["등락률"] = apple_df["등락률"].fillna(0).round(2)

apple_df
```

```python
# 9. 데이터 시각화 >> 시가 / 종가 / 등락폭 / 볼륨 그래프 4개를 subplot 나눠서 4개 표현

import matplotlib.pyplot as plt

plt.subplot(221)
plt.plot(apple_df['Date'],apple_df['Open'])

plt.subplot(222)
plt.plot(apple_df['Date'],apple_df['Close'])

plt.subplot(223)
plt.plot(apple_df['Date'],apple_df['등락폭(종가-시가)'])

plt.subplot(224)
plt.plot(apple_df['Date'],apple_df['Volume'])

plt.show()
```
