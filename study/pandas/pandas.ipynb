```python
# 2022/03/02

import pandas as pd

corona_df = pd.read_csv('corona.csv')

corona_df
```

```python
# column 값 변경

corona_df.columns = ['인덱스', '등록일시', '사망자', '확진자', '게시글번호', '기준일', '기준시간', '수정일시', '누적_의심자', '누적_확진율']

corona_df
```

```python
corona_df.describe()

# 1. count : 컬럼별 데이터의 수
# 2. mean : 컬럼별 데이터의 평균
# 3. std : 컬럼별 데이터의 표준편차
# 4. min : 컬럼별 데이터의 최소값
# 5. max : 컬럼별 데이터의 최대값
# 6. 4분위수 : 컬럼별 25%, 50%, 75%
```

```python
# 차이점이 무엇인가?
# string

corona_df.사망자

corona_df["사망자"]
```

```python
corona_df_2 = corona_df.sort_values("등록일시")

corona_df_2
```

```python
# shift() : column 값을 하나 내리는 함수
# shift(N) : N개만큼 column을 내림

# fillna() : 값이 NaN인 경우 ()값으로 대체
# NaN : Not a Number / 결측값

corona_df_2["일일_확진자"] = (corona_df_2["확진자"] - corona_df_2["확진자"].shift()).fillna(0)
corona_df_2["일일_사망자"] = (corona_df_2["사망자"] - corona_df_2["사망자"].shift()).fillna(0)

corona_df_2
```

```python
# diff()

corona_df_2["일일_사망자"] = corona_df_2["사망자"].diff().fillna(0)

corona_df_2
```

```python
# value_counts() : 컬럼 값의 분포도를 나타냄
# ascending을 통해 count를 기준으로 오름차순, 내림차순을 나타낼 수 있다.

corona_df_2["일일_사망자"].value_counts(ascending=True)
```

```python
# 컬럼을 입력할 때 리스트 형태 [], 기억하기!
# 컬럼명, axis = 1, 축 설정!
# 인덱스명, axis = 0, 축 설정!

corona_df_3 = corona_df_2.drop(['인덱스','기준일','기준시간','수정일시','게시글번호'], axis=1)

corona_df_3
```

```python
# index 초기화 / index 재처리
# drop=True >> 기존 index 삭제
# drop=Flase >> 기존에 index 유지
# inplace = True / 원본값 대체
# inplace = False / 원본값 유지

corona_df_3.reset_index(drop=True, inplace=True)
```

```python
# 결측치 0으로 출력

corona_df_3.fillna(0)
```

```python
# 뒤에서부터 50개 출력

corona_df_4 = corona_df_3.tail(50)

corona_df_4
```

```python
# ffill : 결측치를 앞 데이터로 채우기
# bfill : 결측치를 뒤 데이터로 채우기

corona_df_4.fillna(method='ffill')

corona_df_4.fillna(method='bfill')
```

```python
# 결측치를 평균으로 채우기

corona_df_4.fillna(corona_df_3.mean())
```
